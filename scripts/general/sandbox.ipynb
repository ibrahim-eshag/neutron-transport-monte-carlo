```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
file_path = Path(
    r"C:\Users\Niels\Desktop\nuclear-rust\results\diagnostics\runs\Overnight run\mean_free_path.csv"
)

df = pd.read_csv(file_path)
```

```python
bin_counts, _, _ = plt.hist(df["mean_free_path"], bins=100)

plt.title("Cruciform assembly mean free path")
plt.xlabel("Distance travelled (m)")
plt.ylabel("Neutron count")

mean_free_path = df["mean_free_path"].mean()
plt.vlines(
    x=mean_free_path, ymin=0, ymax=bin_counts.max(), color="tab:green", linestyles="--"
)

plt.savefig(
    "../../figures/10062025 - Mean free path for neutrons in cruciform assembly.png",
    dpi=300,
)
```
